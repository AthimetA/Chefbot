```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = ['Neural Network','Bi-LSTM','Hybrid (CNN-LSTM)','Hierarchical Hybrid \n(CNN-LSTM)','Hierarchical Hybrid \n(CNN-LSTM) with Main class input']
acc       = [0.16,0.36,0.40,0.43,0.59]
precision = [0.22,0.36,0.47,0.40,0.64]
recall    = [0.16,0.36,0.40,0.44,0.59]
f1        = [0.16,0.35,0.40,0.43,0.56]
X_axis = np.arange(len(X))

fig, ax = plt.subplots(figsize=(12,8))

bar_acc = ax.bar(X_axis - 0.3, acc, 0.2, label = 'Accuracy', color='tab:blue')

bar_f1 = ax.bar(X_axis - 0.1, f1, 0.2, label = 'F1-score', color='tab:orange')

bar_precision = ax.bar(X_axis + 0.1, precision, 0.2, label = 'Precision', color='tab:gray')

bar_recall = ax.bar(X_axis + 0.3, recall, 0.2, label = 'Recall', color='tab:green')

for bar in ax.containers:
    ax.bar_label(bar, label_type='edge', fontsize=12)
    

plt.xticks(X_axis, X)
plt.xlabel("Models")
plt.ylabel("Performance")
plt.title("Model Performance")
plt.legend()
plt.show()
```
